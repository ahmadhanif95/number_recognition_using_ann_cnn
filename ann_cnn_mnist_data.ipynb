Importing Libraries

```python
# load all require packages.
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.layers import Flatten, Dense
import matplotlib.pyplot as plt
%matplotlib inline
```

Traning data set preprocessing

```python
train = pd.read_csv("/content/mnist_train.csv")
```

```python
train
```

```python
train_images = train.iloc[:,1:]/255
train_images.head()
```

```python
train_label = train.label
train_label
```

```python
train_image = train_images.values.reshape(train_images.shape[0], 28, 28)
train_image = tf.convert_to_tensor(train_image)
```

```python
train_image
```

Test data set preprocessing

```python
test = pd.read_csv("/content/mnist_test.csv")
```

```python
test
```

```python
test_image = test.drop(columns='label')
```

```python
test_labels= test.label
```

```python
test_labels
```

```python
test_image.shape
```

```python
test_image = test_image.values.reshape(test_image.shape[0],28,28)
```

```python
test_image = tf.convert_to_tensor(test_image)
```

```python
test_image
```

**Multi-Layer Perceptron (ANN) Model**

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(units=200, activation='relu'))
model.add(Dense(units=10, activation='softmax'))
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)
```

```python
model.fit(train_image, train_label, epochs=10)
```

```python
test_loss, test_acc = model.evaluate(test_image, test_labels)
print('Accuracy of model is: {}%'.format(round(test_acc*100, 2)))
```

```python
cnn = models.Sequential([
    layers.Reshape((28, 28, 1), input_shape=(28, 28)),  # Add a reshape layer to include the number of channels
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(units=64, activation='relu'),
    layers.Dense(units=10, activation='softmax')
])

# Rest of your code...

cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
cnn.fit(train_image, train_label, epochs=5)
```

```python
test_loss, test_acc = cnn.evaluate(test_image, test_labels)
print('Accuracy of model is: {}%'.format(round(test_acc*100, 2)))
```
